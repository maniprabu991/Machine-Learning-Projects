```python
#external libraries or modules  allows to use functions and objects defined in those libraries.
import pandas as pd
```

```python
#This is the variable that will store the data read from the CSV file. 
dataset = pd.read_csv("Salary_Dataset.csv")
dataset
```

```python
# It is referred to as "independent" because it is often considered as the input 
independent = dataset[["YEAR OF EXPERIENCE"]]
#It is referred to as "dependent" because it is typically the output or target variable 
dependent = dataset[["SALARY"]]
```

```python
independent
```

```python
dependent
```

```python
# code uses the train_test_split function from the sklearn.model_selection 
# module to split the data into training and testing sets
from sklearn.model_selection import train_test_split
# This divided into four subsets: training features (X_train), 
# testing features (X_test), training targets (Y_train), and testing targets (Y_test).
# it's set to 30%, meaning 70% of the data will be used for training.
X_train,X_test,Y_train,Y_test = train_test_split(independent, dependent, test_size= 0.30, random_state=0)
```

```python
# To create a linear regression model and fits it to the training data.
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train,Y_train)
```

```python
weight = regressor.coef_
weight
```

```python
bias = regressor. intercept_
bias
```

```python
# to predict the dependent variable and value store in variable
Y_predict = regressor.predict(X_test)
#calculate the R-squared score
from sklearn.metrics import r2_score
#To indicates a better fit or not.
r_score = r2_score(Y_test,Y_predict)
```

```python
r_score
```

```python
#defines a filename for saving purpose
import pickle
#This line defines a filename for saving the serialized model .sav commonly used for saved models
filename = "Finalized_Linear_Model.sav"
```

```python
#To save the trained linear regression model
pickle.dump(regressor, open(filename, 'wb'))
```

```python
#The loaded model to make predictions, opens the file and read as binary file
loaded_model = pickle.load(open("Finalized_Linear_Model.sav",'rb'))
#To make a prediction for a new input.
result = loaded_model.predict([[15]])
```
